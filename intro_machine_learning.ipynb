# 1er Modèle : une regréssion linéaire simple

> An algorithm is the general approach you take, the process to construct the model. The model is what you get when you run the algorithm over your training data and what you use to make predictions on new data.

## regression linéaire (simple):
y = Beta0 + Beta1*x 

## regression linéaire (multiple):
Y(vect) = Beta(vect).X(vect) =  Beta0*X0(const) + Beta1*X1 + Beta2*X2 + Beta3*X3 +....

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train) # OLS = Ordinary least squares 
```

```python
y_pred = lm.predict(X_test)
y_pred
```

```python
## suivant juste un axe X 
plt.figure(figsize=(6,2))
plt.scatter(x=X_test[:, 1], y= y_test)
plt.scatter(x=X_test[:, 1], y=lm.predict(X_test), color='r')
```

#### Nous savons ce que le correct output AURAIT DÛ ÊTRE ! 
## que faisons nous dans ce cas ? 

*  Nous choisisons une fonction de coût, 
* c'est à dire une fonction de pénalité L(Y, f(X)) 
* une fonction qui pénalise les erreurs faites sur les predictions.
* "The loss function is how you're penalizing your output."
* il en existe plusieurs avec des normes associées: L1 and L2, Huber loss, Quantile loss, log cosh loss.
* appliquée sur l'ensemble des erreurs faites sur les prédictions: elle retourne un coût : nombre réel qui quantifiera la performance de notre modèle en comparant avec d'autres

Sklearn nous procure aussi un score (R^2 coefficient of determination) pour les régressions, qui quantifient le ratio de la partie de variance expliquée sur la variance totale

```python
lm.score(X_test, y_test)
```

Bon score ! 

# RSS et MSE

```python
RSS_train = sum( (lm.predict(X_train) - y_train)**2 )
RSS_test =  sum( (y_pred - y_test)**2 )

MSE_train, MSE_test =\
    1/X_train.shape[0] * RSS_train, 1/X_test.shape[0]*RSS_test
```

### le coût sur le test est légèrement plus élevé que le train, ce n'est pas illogique...

```python
MSE_train, MSE_test
```

```python
RSS_train, RSS_test
```

## retrouvons le R2 à partir du RSS

Pour ceci on a besoin du SST, et comprendre la part de variance expliquée par le modèle par rapport à celle expliquée par le modèle le plus simple 

```python
SST_test = sum( (np.mean(y_test) - y_test)**2 )
```

```python
R2 = 1 - RSS_test / SST_test
R2
```

## 2ème algorithme, arbre de régression

Commençons par un exemple avec une seule feature X

```python
x = np.arange(0,50)
y1 = np.random.uniform(10,15,10)
y2 = np.random.uniform(20,25,10)
y3 = np.random.uniform(0,5,10)
y4 = np.random.uniform(30,32,10)
y5 = np.random.uniform(13,17,10)

y = np.concatenate((y1,y2,y3,y4,y5))

plt.figure(figsize=(10,4))
plt.scatter(x,y)
```

```python
from sklearn import tree
```

```python
#!pip install colour
from colour import Color
red = Color("red")
colors = list(red.range_to(Color("green"),50))
```

```python
fig, ax = plt.subplots(figsize=(11,8))
for i in [1, 4, 16, 32]:
    clf = tree.DecisionTreeRegressor(max_depth=i)
    model = clf.fit(x[..., np.newaxis],y)
    pred = model.predict(x[..., np.newaxis])
    ax.plot(x, pred, color=colors[i-1].get_rgb(), label='max_depth : {}'.format(i))
ax.scatter(x,y)
ax.legend(loc="right")
```

si on l'avait splitté en test / train ? 

faisons l'opération dans le cas de notre dataset Boston

```python
X_train, X_test, y_train, y_test  = train_test_split(x, y, train_size=0.7)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE_train = {}
MSE_test  = {}
for i in range(1,40):
    clf = tree.DecisionTreeRegressor(max_depth=i)
    model = clf.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    MSE_train[i] = mean_squared_error(y_train, y_pred_train)
    MSE_test[i]  = mean_squared_error(y_test, y_pred)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
pd.Series(MSE_test).plot(kind='line', ax=ax, label="mse_test")
pd.Series(MSE_train).plot(kind='line', color='red', ax=ax, label="mse_train")
ax.legend()
ax.set_title("MSE au cours du temps pour train test")
```

### k-fold cross validation plutôt que one hold-out test set

k-fold is a popular kind of cross-validation technique, in which, say k=10 for example, 9 folds for training and 1 fold for testing purpose and this repeats unless all folds get a chance to be the test set one by one. This way, it provides a good idea of the generalization ability of the model, especially when we have limited data and can't afford to split into test and training data.
